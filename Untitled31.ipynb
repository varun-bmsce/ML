```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

raw_mail_data = pd.read_csv('/content/emails.csv')

mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')

X = mail_data['text']
Y = mail_data['spam']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')

model = LogisticRegression()
model.fit(X_train_features, Y_train)

prediction_on_test_data = model.predict(X_test_features)
accuracy = accuracy_score(Y_test, prediction_on_test_data)

print(f"Accuracy on test data: {accuracy * 100:.2f}%")

input_mail = ["WINNER! You have won a 1000 cash prize. Call now to claim!"]
input_data_features = feature_extraction.transform(input_mail)
prediction = model.predict(input_data_features)

if prediction[0] == 1:
    print("Output: This is a HAM (Not Spam) mail")
else:
    print("Output: This is a SPAM mail")

accuracy = accuracy_score(Y_test, prediction_on_test_data)
precision = precision_score(Y_test, prediction_on_test_data)
recall = recall_score(Y_test, prediction_on_test_data)
f1 = f1_score(Y_test, prediction_on_test_data)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

print("\nClassification Report:")
print(classification_report(Y_test, prediction_on_test_data, target_names=['Ham', 'Spam']))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

raw_mail_data_svm = pd.read_csv('/content/emails.csv')
mail_data_svm = raw_mail_data_svm.where((pd.notnull(raw_mail_data_svm)), '')

X_svm = mail_data_svm['text']
Y_svm = mail_data_svm['spam']

X_train_svm, X_test_svm, Y_train_svm, Y_test_svm = train_test_split(X_svm, Y_svm, test_size=0.2, random_state=42)

feature_extraction_svm = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features_svm = feature_extraction_svm.fit_transform(X_train_svm)
X_test_features_svm = feature_extraction_svm.transform(X_test_svm)

Y_train_svm = Y_train_svm.astype('int')
Y_test_svm = Y_test_svm.astype('int')

model_svm = SVC(kernel='linear', random_state=42)
model_svm.fit(X_train_features_svm, Y_train_svm)

prediction_on_test_data_svm = model_svm.predict(X_test_features_svm)

accuracy_svm = accuracy_score(Y_test_svm, prediction_on_test_data_svm)
precision_svm = precision_score(Y_test_svm, prediction_on_test_data_svm)
recall_svm = recall_score(Y_test_svm, prediction_on_test_data_svm)
f1_svm = f1_score(Y_test_svm, prediction_on_test_data_svm)

print("--- SVM Classifier Results ---")
print(f"Accuracy: {accuracy_svm:.4f}")
print(f"Precision: {precision_svm:.4f}")
print(f"Recall: {recall_svm:.4f}")
print(f"F1 Score: {f1_svm:.4f}")

print("\nClassification Report (SVM):")
print(classification_report(Y_test_svm, prediction_on_test_data_svm, target_names=['Ham', 'Spam']))

input_mail_svm = ["WINNER! You have won a 1000 cash prize. Call now to claim!"]
input_data_features_svm = feature_extraction_svm.transform(input_mail_svm)
prediction_svm = model_svm.predict(input_data_features_svm)

print("\n--- SVM Prediction on New Email ---")
if prediction_svm[0] == 1:
    print("Output: This is a SPAM mail")
else:
    print("Output: This is a HAM (Not Spam) mail")
```
